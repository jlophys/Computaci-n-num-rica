```python
import matplotlib.pyplot as plt
from sympy.solvers import solve
from sympy import*
import numpy as np
```

```python
y = Symbol('y')
f = exp(y) - pi*y
plot(f,(y,0,2))
```

```python
#Calcular la raiz real de la ecuación f(x)=xe^x - pi = 0. Suponer que se desea que el error sea menor que e=0.000001
def f(x): return exp(x) - pi*x
a=1
b=2
e=0.001
r=0
S=[a]
Error=[1]
relativo = [np.absolute((b-a)/(b))]
Conv = []
while relativo[r]>=e:
    r = r+1
    c=(a+b)/2
    print("La solución aproximada en la i-esima iteración es x=",c)
    if f(c)==0:
        print("La solución obtiene en x=",c)
    else:
        if f(a)*f(c)>0:
            a=c
        else:
            b=c
    Anterior = r-1
    S.append(c)
    absoluto = np.absolute(S[r]-S[Anterior])
    rela = np.absolute((S[r]-S[Anterior])/S[r])
    relativo.append(rela)
    Error.append(absoluto)
    Razon = (Error[r]/Error[1])
    Conv.append(Razon)
    print("Error absoluto =",absoluto)
    print("Error relativo =",relativo[r])
    print("Razón convergencia =",Conv)
```

```python
plt.plot(Error)
plt.plot(relativo)
plt.plot(S)
```

```python
t = Symbol('t')
g = exp(t) - pi*t
plot(g,(t,0,2))
```

```python
e=0.001
c = 0
r=0.5
S=[r]
Error=[1]
relativoN = [1]
df = diff(g,t)
while relativoN[c]>=e:
    c = c + 1
    r = r- float(g.subs(t,r))/float(df.subs(t,r))
    S.append(r)
    print(r)
    Anterior = c-1
    absoluto = np.absolute(S[c]-S[Anterior])
    rela = np.absolute((S[c]-S[Anterior])/S[c])
    relativoN.append(rela)
    Error.append(absoluto)
    print("Error absoluto =",absoluto)
    print("Error relativo =",rela)
print(S)
```
